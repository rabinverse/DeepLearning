| PyTorch dtype                    | Description                       | Typical use cases                                                                                        |
| -------------------------------- | --------------------------------- | -------------------------------------------------------------------------------------------------------- |
| `torch.float32` (`torch.float`)  | 32-bit floating point             | Default for most ML/DL models. Training neural networks, general numeric computation.                    |
| `torch.float64` (`torch.double`) | 64-bit floating point             | High-precision math, scientific computing, debugging numerical stability issues. Slower.                 |
| `torch.float16` (`torch.half`)   | 16-bit floating point             | Mixed-precision training on GPUs with Tensor Cores. Reduces memory and speeds up training. Not for CPUs. |
| `torch.bfloat16`                 | 16-bit float with larger exponent | TPUs/Ampere+ GPUs. More stable than fp16, good for mixed precision.                                      |
| `torch.int64` (`torch.long`)     | 64-bit integer                    | **Indices**, **class labels** in classification tasks, embeddings index.                                 |
| `torch.int32` (`torch.int`)      | 32-bit integer                    | Numeric operations requiring integers but not for indices typically.                                     |
| `torch.int16` (`torch.short`)    | 16-bit integer                    | Rarely used, small integer tensors.                                                                      |
| `torch.int8`                     | 8-bit integer                     | Quantized models, memory-efficient storage.                                                              |
| `torch.uint8`                    | Unsigned 8-bit integer            | Images (0–255), masks, binary flags.                                                                     |
| `torch.bool`                     | Boolean                           | Masks, attention masks, condition filtering.                                                             |
| `torch.complex64`                | Complex numbers (2×32-bit floats) | Fourier transforms, signal processing, physics simulations.                                              |
| `torch.complex128`               | Higher precision complex          | High-precision scientific computations.                                                                  |
